```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from tqdm.notebook import tqdm
```

```python
def import_data(file_path):
    data = []
    with open(file_path, 'r') as file:
        lines = file.readlines()
        for line in tqdm(lines[9:]):
            if line.strip():
                columns = line.split()
                if len(columns)>4:
                    timestamp = float(columns[0])
                    trg_id = int(columns[1])
                    brd = int(columns[2])
                    ch = int(columns[3])
                    lg = int(columns[4])
                    hg = int(columns[5])
                else:
                    brd = int(columns[2-2])
                    ch = int(columns[3-2])
                    lg = int(columns[4-2])
                    hg = int(columns[5-2])
                data.append((timestamp, trg_id, brd, ch, lg, hg))
    return pd.DataFrame(data, columns=['Tstamp_us', 'TrgID', 'Brd', 'Ch', 'LG', 'HG'])
```

```python
# Definizione delle funzioni di gaussiana
def gaussian(x, mu, sigma, amplitude):
    return amplitude * np.exp(-(x - mu)**2 / (2 * sigma**2))

def multiple_gaussian(x, *params):
    num_gaussians = len(params) // 3
    result = np.zeros_like(x)
    for i in range(num_gaussians):
        mu = params[i*3]
        sigma = params[i*3 + 1]
        amplitude = params[i*3 + 2]
        result += gaussian(x, mu, sigma, amplitude)
    return result
```

```python
# Definizione dei parametri iniziali per il fit
num_gaussians = 5  # Inserisci il numero di gaussiane che vuoi fittare
initial_params = []
mui=[1500, 1700, 1800, 2000, 1700]
sigmai=[20, 20, 20, 20, 2000]
amplitudei=[400]*5
for i in range(num_gaussians):
    initial_params.extend([mui[i], sigmai[i], amplitudei[i]])  
    #initial_params.extend([mu_guess, sigma_guess, amplitude_guess])  
    # Inserisci le tue stime iniziali per mu, sigma e ampiezza
initial_params
#initial_params=initial_params[:6]
```
